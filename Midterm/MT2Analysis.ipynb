```python
import seaborn as sns 

import matplotlib.pyplot as plt

#raw grades from midterm 1 out of 53 total points
grades01=[31.5,50,43,18,33.5,46,41,14,5,28,45,41,29,39.5,26,53,45,35,24,27,31,14,25,24,41,16,32,21,8,17,46.5,33.5,44,14,20,26,33.5,50,37,48.5,43.5, 52, 48.5]
grades02=[21,24.5,46,29,27,29,42.5,25,28,15,8,50,6,37,37,48,38.5,42,34,37,37,53,35,31,44,38.5,39,38,43.5,29,28,21,27,24.5,43,50,20]

combined = grades01+grades02

sns.set_palette('Set3')
```

```python
#plot the raw scores

fig, ax = plt.subplots(1, 3, figsize=(27, 6))

ax[0].set_title("Section 01 MT 1 Scores /52")
ax[1].set_title("Section 02 MT 1 Scores /52")
ax[2].set_title("Combined MT Scores /52")

sns.histplot(grades01, bins=20, kde=True, ax=ax[0])
sns.histplot(grades02, bins=20, kde=True, ax=ax[1])
sns.histplot(combined, bins=20, kde=True, ax=ax[2])
```

```python
# print the mean, median and 75th percentile in percentage grade (/100)

import statistics 

print("Midterm Averages")
print("================")
print(f"Section 01 Average: {round(sum(grades01)/len(grades01)/52*100,1)}")
print(f"Section 02 Average: {round(sum(grades02)/len(grades02)/52*100,1)}")
print(f"CS 1910 Combined Average: {round(sum(combined)/len(combined)/52*100,1)}")


print("\nMidterm Medians")
print("================")
print(f"Section 01 Median: {round(statistics.median(grades01)/52*100,1)}")
print(f"Section 02 Median: {round(statistics.median(grades02)/52*100,1)}")
print(f"CS 1910 Combined Median: {round(statistics.median(combined)/52*100,1)}")


print("\nMidterm Percentiles")
print("===================")  
print(f"Section 01 75th Percentile: {round(statistics.quantiles(grades01, n=4)[2]/52*100,1)}")
print(f"Section 02 75th Percentile: {round(statistics.quantiles(grades02, n=4)[2]/52*100,1)}")
print(f"Combined 75th Percentile: {round(statistics.quantiles(combined,n=4)[2]/52*100,1)}")
```
