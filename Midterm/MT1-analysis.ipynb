# Midterm Analysis

```python
import seaborn as sns 

import matplotlib.pyplot as plt

#raw grades from midterm 1 out of 53 total points
grades01 = [34.5,40,52,5,49,33.5,41.5,50,33,28,16,41,42,43,49,45,42,48,44,50.5,46,33,12.5,29,32.5,48.5,31,24,28.5,30,16,31,30.5,39,42.5,22,27,31,25,51,48,52,48,46]
grades02 = [26.5,42,38,34,13.5,25,27,26,17.6,23.5,50.5,19,51,38.5,41,37.5,33,51,46,48,49,37,41,42,47,43,37,49,40,35.5,44,49,46.5,35,31.5,33,37,32,40,50,37.5]

combined = grades01+grades02

sns.set_palette('Set3')
```

```python
#plot the raw scores

fig, ax = plt.subplots(1, 3, figsize=(27, 6))

ax[0].set_title("Section 01 MT 1 Scores /53")
ax[1].set_title("Section 02 MT 1 Scores /53")
ax[2].set_title("Combined MT Scores /53")

sns.histplot(grades01, bins=20, kde=True, ax=ax[0])
sns.histplot(grades02, bins=20, kde=True, ax=ax[1])
sns.histplot(combined, bins=20, kde=True, ax=ax[2])
```

```python
# print the mean, median and 75th percentile in percentage grade (/100)

import statistics 

print("Midterm Averages")
print("================")
print(f"Section 01 Average: {round(sum(grades01)/len(grades01)/53*100,1)}")
print(f"Section 02 Average: {round(sum(grades02)/len(grades02)/53*100,1)}")
print(f"CS 1910 Combined Average: {round(sum(combined)/len(combined)/53*100,1)}")


print("\nMidterm Medians")
print("================")
print(f"Section 01 Median: {round(statistics.median(grades01)/53*100,1)}")
print(f"Section 02 Median: {round(statistics.median(grades02)/53*100,1)}")
print(f"CS 1910 Combined Median: {round(statistics.median(combined)/53*100,1)}")


print("\nMidterm Percentiles")
print("===================")  
print(f"Section 01 75th Percentile: {round(statistics.quantiles(grades01, n=4)[2]/53*100,1)}")
print(f"Section 02 75th Percentile: {round(statistics.quantiles(grades02, n=4)[2]/53*100,1)}")
print(f"Combined 75th Percentile: {round(statistics.quantiles(combined,n=4)[2]/53*100,1)}")
```
