```python
# Accessing Values

student = {
    "name": "Alice",
    "age": 75,
    "address": "Charlottetown"
}

# Accessing values
print(f"Student's name: {student['name']}")
print(f"Student's age: {student['age']}")

# Trying to access a non-existent key 
#(will cause an error if uncommented)
# print(student["city"])
```

```python
#modifying dictionaries

student = {
    "name": "Alice",
    "age": 75,
    "address": "Charlottetown"
}

print(student)

#update the student's address
student["address"] = "Morell"
print(student)

#add a new key-value pair
student["GPA"] = 4.0

print(student)

#remove a key-value pair

#del
del student["age"]
print(student)

print(student.pop("address"))

print(student)
```

```python
# .pop
student = {
    "name": "Alice",
    "age": 75,
    "address": "Charlottetown"
}

age = student.pop("age")
print(age)

#pop GPA but supply a default
#the default is returned if GPA is 
#not a valid key in the dictionary
gpa = student.pop("GPA", 0.0)
print(gpa)

print(student)
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Create a view object for keys
keys_view = my_dict.keys()
print(f"Initial keys view: {keys_view}")

# Modify the dictionary
my_dict['d'] = 4
print(f"Keys view after dictionary modification: {keys_view}")

```

```python
# Iterating over dictionaries
my_dict = {'a': 1, 'c': 3, 'b': 2}

my_dict['d'] = 4

#iterates over the keys 
#(in the order they were inserted)
print("iterate over keys")

for k in my_dict:
    print(k)

#iterates over the keys
for k in my_dict.keys():
    print(k)

print("iterate over values")
#iterates over the values 
for v in my_dict.values():
    print(v)

print("iterate over items")
for k,v in my_dict.items():
    print(k,v)

for item in my_dict.items():
    print(item[0], item[1])


#iterate over the keys but access the values
print("iterate keys but access value")
for k in my_dict:
    print(k, my_dict[k])
```

```python
# class participation
#Which approach adds 10 to each value in a dictionary

my_dict={ 'a':1, 'b':2, 'c':3}

#A
for v in my_dict.values():
    v+=10
#B
for k in my_dict:
    my_dict[k] += 10

#C
for k,v in my_dict.items():
    my_dict[k] = v + 10
#D
my_dict = {k:v+10 for k,v in my_dict.items()}

print(my_dict)
```
