```python
# Deck of cards example

# highest to lowest rankings
suit = ["♠", "♥", "♦", "♣"]
rank = ["A", "K", "Q", "J", "10", "9", "8", "7", "6", "5", "4", "3", "2"]

#10 of hearts
ten_hearts = rank[4] + suit[1]
print(f"10 of hearts: {ten_hearts}")

# create a deck of cards with all cards grouped by suit and ordered by rank
# the highest ranking suits should be ordered first

deck = []

for s in suit:
    for card in rank:
        deck.append(card + s)

print(deck)

# suppose you want to swap the positions of the 2 of hearts and the 2 of clubs?
idx2h = deck.index("2♥")
idx2c = deck.index("2♣")

#swap method 1
deck[idx2h], deck[idx2c] = deck[idx2c], deck[idx2h]

#swap method 2
#tmp = deck[idx2h]
#deck[idx2h] = deck[idx2c]
#deck[idx2c] = tmp

print(deck)
```

```python
# visualizing data with sound 

# find a partner

# the below code generates various waveforms (sinewave, triangle wave, sawtooth wave)
# your task is to build a square wave 
# square wave is -1 for some number of samples and then +1 for the same number of samples repeating. 

import numpy as np
from IPython.display import Audio, display
import math



#samples per second of audio
sample_rate = 44100


sine_wave = []
sawtooth_wave = []
triangle_wave = []
sq_wave = []

# 44100 samples is one second of audio
for i in range(sample_rate):
    sine_wave.append(math.sin(i / 12))
    sawtooth_wave.append((i % 200) / 100 - 1)
    triangle_wave.append( ((i % 200) - (i % 200) // 100 * (2*i % 200)) / 50 - 1)
    
    # alter this line to create the square wave
    sq_wave.append( 0 )

print(f"saw_tooth_wave, freq = {44100 / 200}")
display(Audio(sawtooth_wave, rate=sample_rate))

print(f"sine wave freq={44100 / 12 / (2 * math.pi)}")
display(Audio(sine_wave, rate = sample_rate))

print(f"triangle wave freq={44100 / 200}")
display(Audio(triangle_wave, rate = sample_rate))

#alter this line to output the frequency
print(f"square wave freq={44100 / }")
display(Audio(sq_wave, rate = sample_rate))


#compare against: https://musiclab.chromeexperiments.com/Oscillators/
```

```python
# Immutable object assignment

#ints are immutable 
value = 7

cpy_value = value

#this creates a new integer that is 10 greater than cpy_value (and then it is referenced by the cpy_value variable)
cpy_value += 10 

#value is unchanged
print(value)
```

```python
# Mutable Object Assignment

lst1 = [1, 2, 3]

# assign lst2 to reference whatever lst1 references
lst2 = lst1

#change lst1
lst1.append(20) 

# the 20 also prints out 
print(lst2) #[1, 2, 3, 20]

lst2[1] = 7

# also 7 in index 1
print(lst1) #[1, 7, 3, 20]

#but reassignment is different
lst2 = [8, 9, 10] #point lst2 somewhere else (not a modifying operation)

print(lst1) #still [1, 7, 3, 20]
```

```python
# example of a function with a side effect
# it modifies the argument passed in

def modifying_function(x):
    """ modify the parameter 
    
    """
    x[1] = 7


my_lst = [1, 2, 4]
modifying_function(my_lst)

print(my_lst) # [1, 7, 4]
```

```python
# copying lists

original_list = [1, 2, 3]

#.copy
cpy = original_list.copy() 

cpy[2] = 29 #original is unaffected

#slicing
cpy2 = original_list[:]

#list constructor
cpy3 = list(original_list)

#copy.copy
import copy
cpy4 = copy.copy(original_list)

cpy4.append(76) #only cpy4 is affected

print(original_list)

```

```python
# deep copy
import copy

orig = [1, 2, [1, 2, 3]]

shallow = orig.copy()

#alter the mutable object in the copy 
shallow[2].append(600)

#ack the original is changed too
print(orig)

cpy = copy.deepcopy(orig)
cpy[2].append(860)

#not affected 
print(orig)
```

```python
# concatenate lists

lst1 = [1, 2, 3]
lst2 = [2, 4, 6]

# + operator
lst3 = lst1 + lst2

# += operator
lst1 += lst2 

lst2[0] = 88
lst3[1] = 29

print(lst1)

```

```python
# unpacking

lst = [1, 2, 3]

#lst is nested inside lst2
lst2 = [lst, 5, 6]

print(lst2) #[[1, 2, 3], 5, 6]

#the elements of lst are unpacked inside lst2
lst3 = [*lst, 5, 6]

print(lst3) #[1, 2, 3, 5, 6]
```
