````python
# L15 Examples - End of Lists, List Comprehensions, Multidimensional Lists, Tuples

# List comprehensions

#existing list
numbers = [1, 2, 3, 4, 5]

#use a list comprehension to generate squares of the original values 
squares = [x**2 for x in numbers]
# squares will be [1, 4, 9, 16, 25]


#existing list
words = ["apple", "banana", "cherry", "date"]

#list of long words from the existing list
long_words = [word for word in words if len(word) > 5]
# long_words will be ["banana", "cherry"]
````

````python
# without list comprehensions

# square of numbers
sq_nums = []
for num in numbers:
    sq_nums.append(num**2)


#long words 
long_wrds = []
for word in words:
    if len(word) > 5:
        long_wrds.append(word)


````

````python
# Nested Lists

lst = [ [1, 2, 3], [4, 5, 6], [7, 8, 9] ]

print(lst) # [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(len(lst)) # 3
print(len(lst[0])) # 3
print(lst[0][2]) # 3
````

````python
lst = [ [1, 2, 3], [5, 4, 6], [7, 8, 9] ]

# accessing individual elements
lst[1][2] # 2nd index position of the 1st element in lst

# elements of lst are lists 
lst[1].sort()
print(lst) # [ [1, 2, 3], [4, 5, 6], [7, 8, 9] ]

lst.sort(reverse=True)
print(lst) # [ [7, 8, 9], [4, 5, 6], [1, 2, 3] ]

#iterating over a multidimensional list

print("for each")
#using for each loop
for sub_lst in lst:
    for value in sub_lst:
        print(value, end = " ")
    print()

print("using indices")
#using indices
for i in range(len(lst)):
    for j in range(len(lst[i])):
        print(lst[i][j], end = " ")
    print()

        

````
