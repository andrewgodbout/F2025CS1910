```python
# The following code will download the required data files for this lecture

from os.path import basename, exists

def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve

        local, _ = urlretrieve(url, filename)
        print("Downloaded " + str(local))
    return filename

examples_path = "https://github.com/andrewgodbout/F2025CS1910/raw/refs/heads/main/lectures/L19/ex.zip"


download(examples_path)
!unzip -o ex.zip
```

```python
# opening, reading and closing files


f = open("temperature.txt", "r")

#reading a file line by line
for line in f:
    print(line)
f.close()
```

```python
# read line by line and store into list

f = open("temperature.txt", "r")

lines = []
for line in f:
    lines.append(line)

f.close()
print(lines)
```

```python
# remove the newlines

f = open("temperature.txt", "r")

lines = []
for line in f:
    # strip away trailing (rightside) whitespace with rstrip
    lines.append(line.rstrip())

f.close()
print(lines)
```

```python
# self-check

#what is the output?
f = open("temperature.txt", "r")

cnt = 0
for temp_line in f:
    if cnt == 0:
        mn_temp = temp_line
    if temp_line < mn_temp:
        mn_temp = temp_line
    cnt += 1
f.close()
print(f"min: {mn_temp}")


```

```python
# converting Text data

f = open("temperature.txt", "r")

cnt = 0
for temp_line in f:
    temp_line = float(temp_line)
    if cnt == 0:
        mn_temp = temp_line
    if temp_line < mn_temp:
        mn_temp = temp_line
    cnt += 1
f.close()
print(f"min: {mn_temp}")
```

```python
# Convert to list

f = open("temperature.txt", "r")

f_list = list(f)
f_list_float = [float(flt) for flt in f_list]

#print the min
print(min(f_list_float))
f.close()
```

```python
#write data to a file

numbers = [42 , 24 , 87 , 21 , 76]

f = open("numbers.txt", "w")

cnt = 0
for n in numbers:
    if cnt < len(numbers)-1:
        f.write(str(n) + ",")
    else:
        f.write(str(n))
    cnt += 1

f.close()
```

```python
# alternative with .join

numbers = [42 , 24 , 87 , 21 , 76]

f = open("numbers.txt", "w")
f.write( ",".join([str(n) for n in numbers])
f.close()
```
