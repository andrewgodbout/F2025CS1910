```python
## Dictionary Comprehensions


numbers = [1, 2, 3, 4]
squared_dict = {num: num**2 for num in numbers}
# Result: {1: 1, 2: 4, 3: 9, 4: 16}
print(squared_dict)

#optional condition
original_dict = {'a': 10, 'b': 25, 'c': 5, 'd': 30}
filtered_and_doubled = {k: v * 2 for k, v in original_dict.items() if v > 15}
# Result: {'b': 50, 'd': 60}

print(filtered_and_doubled)
```

```python
# dictionary from list of tuples

data = [('apple', 1), ('banana', 2), ('cherry', 3)]
fruit_counts = {fruit: count for fruit, count in data}
# Result: {'apple': 1, 'banana': 2, 'cherry': 3}

print(fruit_counts)

#if-else condition
data = {'A': 10, 'B': 5, 'C': 12}
transformed_values = {k: "High" if v > 10 else "Low" for k, v in data.items()}
# Result: {'A': 'High', 'B': 'Low', 'C': 'High'}

print(transformed_values)
```
